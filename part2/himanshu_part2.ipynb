```python
state = ["IA","IC","IG","IL","IM","IS","LE","LG","LI","LL","LT","MA",
         "MG","MH","ML","MN","MP","NC","ND","NE","NG","NL","NO","NP","NR","RE","SP","ST"]
```

```python
x=[]
x=range(1,29)
```

```python
def mark(top):
    for top in top:
        height = top.get_height()
        plt.text(top.get_x()+top.get_width()/6., 1.01*height, '%s' % int(height),fontsize=14)
```

```python
plt.rcParams["figure.figsize"] = (15,15)
top=plt.bar(x,height,align="center",width=0.5,alpha=0.5)
plt.xticks(x,state)
plt.rc('xtick', labelsize = 15)
plt.rc('ytick', labelsize = 15) 
plt.ylabel('Number of grants alloted',fontsize = 20)
plt.title('Programs Type',fontsize = 20)
mark(top)
plt.show()
```

```python
grant = pd.read_csv("DiscGrants96to13.csv", encoding = 'latin')
```

```python
# To plot the lat and long I referred Stackoverflow.
from mpl_toolkits.basemap import Basemap as Basemap
from matplotlib.patches import Polygon
```

```python
plt.rcParams["figure.figsize"] = (10,10)
lon=grant["Longitude"].tolist()
lat=grant["Latitude"].tolist()
map = Basemap(llcrnrlon=-119,llcrnrlat=22,urcrnrlon=-64,urcrnrlat=49,
        projection='lcc',lat_1=32,lat_2=45,lon_0=-95,lat_0=26)
shape_info=map.readshapefile('shape', name='states', drawbounds=True)
x,y = map(lon,lat)
map.plot(x,y,'sg', markersize=4)
plt.show()
```

```python
plt.rcParams["figure.figsize"] = (5,5)
map = Basemap(projection='ortho',lat_0=25,lon_0=-95,resolution='c')
# draw coastlines, country boundaries, fill continents.
map.drawcoastlines(linewidth=0.25)
map.drawcountries(linewidth=0.25)
map.fillcontinents(color='coral',lake_color='aqua')
# draw the edge of the map projection region (the projection limb)
map.drawmapboundary(fill_color='aqua')
map.drawmeridians(np.arange(0,360,30))
map.drawparallels(np.arange(-90,90,30))
x, y = map(lon,lat)
map.plot(x,y)
plt.title('Orthogonal projection of Lat Lon')
plt.show()
```
